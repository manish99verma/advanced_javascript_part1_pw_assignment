##Problem 2
Create a Map to store contact information (name, age, email, location) and implement a function to retrieve contact details by name.
- Contact Manager with map implementation given below:
  ```
  class Contact {
    name;
    age;
    email;
    location;

    constructor(name, age, email, location) {
      this.name = name;
      this.age = age;
      this.email = email;
      this.location = location;
    }
  }

  class ContactManager {
    constructor() {
      this.contacts = new Map();
    }

    addContact(name, age, email, location) {
      this.contacts.set(name, new Contact(name, age, email, location));
    }

    findContact(name) {
      if (this.contacts.has(name)) {
        const foundContact = this.contacts.get(name);
        return { ...foundContact };
      }

      console.log(`${name} does not exists in contact list!`);
    }
  }

  const contactManager = new ContactManager();
  contactManager.addContact(
    "Manish",
    23,
    "manish6203993@gmial.com",
    "Nalanda, Bihar"
  );
  contactManager.addContact("Ramesh", 26, "ramesh@gmail.com", "Delhi, India");
  contactManager.addContact(
    "Suresh",
    22,
    "suresh123@gmail.com",
    "Kolkata, WestBengal"
  );

  console.log(`Find Contact with Manish`, contactManager.findContact("Manish"));
  console.log(`Find Contact with Suresh:`, contactManager.findContact("Suresh"));
  ```
